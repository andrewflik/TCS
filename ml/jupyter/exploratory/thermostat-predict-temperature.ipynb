```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# load dataset
dataset = numpy.genfromtxt("datasets/thermostat/sample-training-data.csv", delimiter=",")
X = dataset[:,0:11].astype(float)
Y = dataset[:,11]
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
print(encoded_Y)
print(dummy_y)
```

```python
input_dimension = X.shape[1]
print(input_dimension)
# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(11, activation='relu', input_dim=input_dimension))
    model.add(Dense(units=3))
    model.add(Activation("softmax"))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = baseline_model()
```

```python
model.fit(X,dummy_y, epochs=100, batch_size=3)
```

```python
def getTestInput():
    month=12
    day=12
    hour=0
    minute=0
    pirstatus=0
    isDay=0
    extTemp=5
    extHumidity=55
    loungeTemp=16
    loungeHumidity=35
    state=0
    
    return numpy.array([month,day,hour,minute,pirstatus,isDay,extTemp,extHumidity,loungeTemp,loungeHumidity,state]).reshape((1,11))
```

```python
#input = numpy.array(X[17]).reshape((1,11))
input  = getTestInput()
print(input)
result = model.predict(input)
for res in result:
    print(res)
result = model.predict_classes(input)
for res in result:
    print(res)
```
